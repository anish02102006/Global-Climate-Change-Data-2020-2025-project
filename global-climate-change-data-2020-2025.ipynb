# Exploratory Data Analysis (EDA)

# Average Temperature Trend

```python
sns.lineplot(data=df, x='Year', y='Avg_Temperature(°C)', hue='Continent', marker='o')
plt.title('Average Temperature Trend (2020–2025)')
plt.ylabel('Avg Temperature (°C)')
plt.show()
```

# CO₂ Emissions Trend 

```python
sns.barplot(data=df, x='Year', y='CO2_Emissions(Mt)', hue='Continent')
plt.title('CO2 Emissions by Continent')
plt.show()
```

# Sea Level Rise 

```python
sns.boxplot(data=df, x='Continent', y='Sea_Level_Rise(mm)')
plt.title('Sea Level Rise Distribution by Continent')
plt.show()
```
